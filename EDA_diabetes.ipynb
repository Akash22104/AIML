```python
#imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
#load datasets
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
diabetes
```

```python
diabetes.target[:3]
```

```python
diabetes.data.shape
```

```python
#Convert to dataframe:
df=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
df["target"]=diabetes.target
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
#check missing values:
print("Missing values:")
df.isnull().sum()
```

```python
#Remove duplicates:
print("No. of duplicate rows:",df.duplicated().sum())
df=df.drop_duplicates()
```

```python
df.shape
```

```python
#Detecting outliers using Percentile:
for	col	in	df.columns:
    Q1	=	df[col].quantile(0.25)
    Q3	=	df[col].quantile(0.75)
    IQR	=	Q3	-	Q1
    lower	=	Q1	-	1.5	*	IQR
    upper	=	Q3	+	1.5	*	IQR
    outliers	=	df[(df[col]	<	lower)	|	(df[col]	>	upper)]
    print(f"{col:8s}	:	{len(outliers)}	outliers")
```

```python
import	matplotlib.pyplot	as	plt
plt.figure(figsize=(8, 6))
plt.hist(df["target"],	bins=25,	color="blue",	edgecolor="black")
plt.title("Distribution	of	Disease	Progression	(Target)")
plt.xlabel("Disease	progression	value")
plt.ylabel("Frequency")
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(7,	4))
plt.scatter(df["bmi"],	df["target"],	alpha=0.6,	color="orange")
plt.title("BMI	vs	Disease	Progression")
plt.xlabel("BMI	(standardized)")
plt.ylabel("Disease	progression")
plt.grid(alpha=0.5)
plt.show()
```

```python
#Box plots:
plt.figure(figsize=(9,	4))
plt.boxplot([df[c] for	c	in	diabetes.feature_names],labels=diabetes.feature_names)
plt.title("Boxplot	of	Diabetes	Dataset	Features")
plt.xlabel("Feature")
plt.ylabel("Standardized	value")
plt.grid(alpha=0.3)
plt.show()
```

```python
correlations=df.corr()["target"].drop("target")
print("Correlation between target and orther columns",correlations)
```

```python
#Correlation Heatmap:
import	seaborn as sns
import	numpy	as	np
corr=df.corr()
plt.figure(figsize=(8,	6))
sns.heatmap(corr,annot=True,cmap="coolwarm",vmin=-1,vmax=1)
plt.title("Correlation	Heatmap	-	Diabetes	Dataset")
plt.show()
```

```python
counts = df["sex"].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(counts,labels=["Male", "female"],autopct="%.2f%%",colors=["red", "pink"],startangle=90,explode=(0.03, 0))
plt.title("Sex Distribution of Patients")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df["s1"], df["s2"], alpha=0.5, color="green")
plt.title("s1 vs s2 ")
plt.xlabel("s1 - total cholesterol")
plt.ylabel("s2 - LDL")
plt.grid(alpha=0.3)
plt.show()
```

```python
groups = [df["target"].values for x in df.groupby("sex")]

plt.figure(figsize=(6, 4))
plt.boxplot(groups, labels=["male", "female"])
plt.title("Disease Progression by Sex")
plt.ylabel("Disease progression")
plt.grid(alpha=0.5)
plt.show()
```
