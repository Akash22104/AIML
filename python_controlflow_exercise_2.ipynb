```python
num = int(input("Enter the number:"))
total = 0

while num > 0:
    total += num % 10
    num //= 10

print(total)
```

```python
#reverse the number:
num = int(input())
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

print(rev)
```

```python
num = int(input())
temp = num
rev = 0

while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10

if rev == num:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#prime numbers betwwen 1 to 50:
for num in range(2, 51):
    prime = True

    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num, end=" ")
```

```python
#Hollow square pattern:
n = int(input("Enter the length of square:"))
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
#Guess the number game:
n=int(input("Enter the secret number:"))
secret=n
while True:
    guess = int(input("Guess the number?"))
    if guess==secret:
        print("Correct guess!!")
        break
    elif guess<secret:
        print("Guess higher")
    else:
        print("Guess Lower")
```

```python
#Rock,paper,scissors game:
player1=input("player 1:").lower()
player2=input("player 2:").lower()

if player1==player2:
    print("Tie")
elif (player1 == "rock" and player2 == "scissors") or \
     (player1 == "paper" and player2 == "rock") or \
     (player1 == "scissors" and player2 == "paper"):
    print("Player 1 Wins")
else:
    print("Player 2 WINS!!")
   
```

```python
##### Simple calculator with menu:

while True:
    print("\n1.Add")
    print("2.Subraction")
    print("3.multiplication")
    print("4.Division")
    print("5.Exist")

    choice = int(input("Enter the choice:"))

    if choice==5:
        print("Existed!")
        break
    a=float(input("Enter first number:"))
    b=float(input("Enter second number:"))

    if choice==1:
        print("Result=",a+b)
    elif choice==2:
        print("Result=",a-b)
    elif choice==3:
        print("Result=",a*b)
    elif choice==4:
        if b==0:
            print("Cannot divide by 0")
        else:
            print("Result=",a/b)

    else:
        print("Invalid choice!")
        
```

```python
password = input("Enter password: ")

has_digit = False
has_upper = False

for ch in password:
    if ch.isdigit():
        has_digit = True
    if ch.isupper():
        has_upper = True

if len(password) >= 8 and has_digit and has_upper:
    print("Valid Password")
else:
    print("Invalid Password")
```

```python
balance = 10000

while True:
    print("\n1.Check Balance")
    print("2.Deposit")
    print("3.Withdraw")
    print("4.Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        print("Balance =", balance)

    elif choice == 2:
        amount = float(input("Enter amount: "))
        balance += amount
        print("Updated Balance =", balance)

    elif choice == 3:
        amount = float(input("Enter amount: "))

        if amount <= balance:
            balance -= amount
            print("Withdrawal Successful")
            print("Remaining Balance =", balance)
        else:
            print("Insufficient Balance")

    elif choice == 4:
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```
