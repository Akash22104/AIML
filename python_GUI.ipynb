```python
import customtkinter as ctk

# Set appearance
ctk.set_appearance_mode("dark")      # Options: "light", "dark", "system"
ctk.set_default_color_theme("blue")  # Themes: "blue", "green", "dark-blue"

# Create main window
app = ctk.CTk()
app.title("Modern Frontend")
app.geometry("600x400")

# Title
title = ctk.CTkLabel(
    app,
    text="Welcome to My Application",
    font=("Arial", 24, "bold")
)
title.pack(pady=20)

# Username Entry
username = ctk.CTkEntry(
    app,
    placeholder_text="Enter Username",
    width=250
)
username.pack(pady=10)

# Password Entry
password = ctk.CTkEntry(
    app,
    placeholder_text="Enter Password",
    show="*",
    width=250
)
password.pack(pady=10)

# Function
def login():
    print("Username:", username.get())
    print("Password:", password.get())

# Login Button
login_btn = ctk.CTkButton(
    app,
    text="Login",
    command=login,
    width=200
)
login_btn.pack(pady=20)

# Checkbox
remember = ctk.CTkCheckBox(
    app,
    text="Remember Me"
)
remember.pack()

# Progress Bar
progress = ctk.CTkProgressBar(app, width=250)
progress.set(0.75)
progress.pack(pady=20)

# Run application
app.mainloop()
```

```python
import tkinter as tk
def show_text():
    entered_text = entry.get()
    label.config(text=f"You typed: {entered_text}")
root = tk.Tk()
root.title("Textbox and Button Example")
entry = tk.Entry(root, width=80)
entry.pack(pady=20)
button = tk.Button(root, text="Submit", command=show_text)
button.pack(pady=20)
label = tk.Label(root, text="")
label.pack(pady=20)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
def show_message():
    messagebox.showinfo("Greetings", "Hello! This is a message box.")
root = tk.Tk()
root.title("Message Box Example")
button = tk.Button(root, width=50, text="Click Me", command=show_message)
button.pack(pady=20)
root.mainloop()
```

```python
import tkinter as tk
def show_choice():
    label.config(text=f"Selected Option: {var.get()}")
root = tk.Tk()
root.title("Radio Button Example")

var = tk.StringVar(value="None")

radio1 = tk.Radiobutton(root, width=50, text="Option 1", variable=var, value="Option 1", command=show_choice)
radio2 = tk.Radiobutton(root, width=50, text="Option 2", variable=var, value="Option 2", command=show_choice)
radio3 = tk.Radiobutton(root, width=50, text="Option 3", variable=var, value="Option 3", command=show_choice)

radio1.pack()
radio2.pack()
radio3.pack()
                        
label = tk.Label(root, text="Selected Option: None")
label.pack(pady=10)
    
root.mainloop()
```
