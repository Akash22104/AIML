```python
#Prove NumPy is fasted than List:
import time

#Creating large list and Array:
python_list=list(range(1000000))
numpy_array=np.arange(1000000)

#Time elements-wise operation
start=time.time()
result_list=[x*2 for x in python_list]
print("Python list time:" ,time.time()-start,"seconds")

start=time.time()
result_array=numpy_array*2
print("NumPy array time:", time.time()-start,"seconds")

#Memory efficiency:
print("Python list size:", python_list.__sizeof__(),"bytes")
print("NumpPy array size:",numpy_array.nbytes,"bytes")
```
