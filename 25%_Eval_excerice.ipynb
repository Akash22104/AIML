```python
#1.Palindrome:
def palindrome(text):
    text=text.lower()

    if text==text[::-1]:
        return True
    else:
        return False
word= input("Enter String:")

if palindrome(word):
    print("Palindrome")
else:
    print("Not a palindrome")
    
```

```python
text = input("Enter a string: ")

frequency = {}

for ch in text:
    frequency[ch] = frequency.get(ch, 0) + 1

counts = list(frequency.values())
counts.sort(reverse=True)

if len(counts) < 2:
    print("Second-highest frequency does not exist")

else:
    highest = counts[0]

    second_highest = 0

    for count in counts:
        if count < highest:
            second_highest = count
            break

    if second_highest == 0:
        print("Second-highest frequency does not exist")
    else:
        print("Character(s) with second-highest frequency:")

        for ch in frequency:
            if frequency[ch] == second_highest:
                print(ch)
                
```

```python
def count_case_letters(text):
    uppercase = 0
    lowercase = 0

    for ch in text:
        if ch.isupper():
            uppercase += 1
        elif ch.islower():
            lowercase += 1

    return uppercase, lowercase


text = input("Enter a string: ")

upper, lower = count_case_letters(text)

print("Uppercase letters =", upper)
print("Lowercase letters =", lower)
```

```python
number = int(input("Enter a number: "))

order = len(str(number))
temp = number
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** order
    temp //= 10

if total == number:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
def merge_dictionaries(dict1, dict2):
    result = dict1.copy()

    for key in dict2:
        result[key] = dict2[key]

    return result


dict1 = {"name": "Akash", "age": 20}
dict2 = {"city": "Chennai", "age": 21}

merged_dict = merge_dictionaries(dict1, dict2)

print(merged_dict)
```

```python
def count_vowel_words(sentence):
    words = sentence.split()
    count = 0

    for word in words:
        if word[0].lower() in "aeiou":
            count += 1

    return count


sentence = input("Enter a sentence: ")

print("Words starting with vowels =", count_vowel_words(sentence))
```
