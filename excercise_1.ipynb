```python
#21.Check if a number is prime:
num=int(input("Enter the number:"))

if num>1:
    for i in range(2,int(num ** 0.5) + 1):
        if num % i == 0:
            print("Not a prime")
            break
    else:
            print("Is a prime")
else:
    print("Not a prime")
```

```python
#25.Check if a number is Armstrong number:
num = int(input("Enter a number: "))

order = len(str(num))
temp = num
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** order #total+di
    temp //= 10

if total == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
#33.Count word frequency in a string using a dictionary:
text=input("Enter the string:")
words = text.split()
freq={}
for word in words:
    freq[word]=freq.get(word,0)+1
print(freq)
```
