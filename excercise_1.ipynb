```python
#6.Reverse a string
text=input("Enter the string to be reversed:")
reversed_text=text[::-1]
print("The Reversed String:",reversed_text)
```

```python
#7.Check if a string is a palindrome
text=input("Enter the String to be checked if palindrome or not:")
if text==text[::-1]:
    print("Given text is palindrome")
else:
    print("Given text is not an palindrome")
```

```python
#8.Count vowels and consonants in a string:
text = input("Enter a string: ").lower()
vowels = 0
consonants = 0
for x in text:
    if x.isalpha():
        if x in "aeiou":
            vowels += 1
        else:
            consonants += 1
print("Vowels:", vowels)
print("Consonants:", consonants)
```

```python
#9.Remove duplicate characters from a string
text = input("Enter the text:").lower()
result=""
for x in text:
    if x not in result:
        result=result+x
print("After removing every duplicate characters:",result)
```

```python
#10.Check if two strings are anagrams:
string1= input("Enter the string 1")
string2=input("Enter the string 2")
if sorted(string1)==sorted(string2):
    print("Given strings are anagrams")
else:
    print("Given strings are not anagrams")
```
