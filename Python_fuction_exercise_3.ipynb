```python
#1.Write a function to return the square of a number:
def square(n):
    return n*n
num=int(input("Enter the number:"))
print("The square of number:",square(num))
```

```python
#2.Write a function that adds two numbers:
def add(a,b):
    return a+b
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
print("Result=", add(a,b))
```

```python
#3.Write a function to check if a number is even:
def even(n):
    return n%2==0
n=int(input("Enter the number:"))
if even(n):
    print("Even")
else:
    print("Odd")
```

```python
#4.Write a function that returns the largest number from a list:
def largest_num(lst):
    return max(lst)
numbers=[10,20,30,40,50,60,70,80,90]
print("The largest number:",largest_num(numbers))
```

```python
#5.Write a function to compute factorial of a number:
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact
    
num=int(input("Enter the number:"))
print("The factorial:", factorial(num))
```

```python
#6.Write a function to reverse a string:
def reverse_string(s):
    return s[::-1]
s=input("Give String:")
print("The reverse string:",reverse_string(s))
    
```

```python
#7.Write a function to check if a string is a palindrome:
def palindrome(s):
    return s==s[::-1]
s=input("Enter the string:")
if palindrome(s):
    print("Is a palindrome")
else:
    print("Not a palindrome")
    
```

```python
#9.Count vowels in a string:
def vowels(s):
    count=0
    for ch in s.lower():
        if ch in "aeious":
            count+=1
    return count
text=input("Enter the string:")
print("Vowels=",vowels(text))
```

```python
#10.write a function that checks if a number is prime:
def prime(n):
    if n<2:
        return False

    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
num=int(input("Enter a number:"))
if prime(num):
    print("Prime")
else:
    print("Not Prime")
```

```python
#11.Generate fibonacci series up to n terms:
def fibonacci(n):
    a=0
    b=1
    for i in range (1,n+1):
        print(a," ")
        a,b=b,a+b
terms=int(input("Enter how many terms:"))
fibonacci(terms)
```

```python
#12.write a function to convert celcius to farenheit:
def celsius_farenheit(c):
    return (c*9/5)+32
celsius=float(input("Enter the Temperature in Celsius:"))
print("The temperature in farenheit:",celsius_farenheit(celsius))
```

```python
#16.Write a function that finds the GCD of two numbers:
def GCD(a,b):
    while b!=0:
        a,b=b,a%b
    return a
x=int(input("Enter first number:"))
y=int(input("Enter second number:"))
print("The GCD=",GCD(x,y))
```

```python
#18.Write a function to count the number of words in a string:
def count_string(string):
    return len(string.split())
sentence=input("Enter the string:")
print("Count of words:",count_string(sentence))
```

```python
#19.Write a function that accepts a number and returns the sum of its digits:
def sum_digits(num):
    total=0

    while num>0:
        total+=num%10
        num//=10
    return total
number=int(input("Enter a number:"))
print("Sum of digits=",sum_digits(number))
```

```python
#20.Write a function to check if a given year is a leap year:
def leap_year(year):
    if (year%4==0 or year%100!=0) or (year%400==0):
        return True
    return False

year=int(input("Enter a year:"))

if leap_year(year):
    print("Is a Leap Year")
else:
    print("Not a leap Year")
```
