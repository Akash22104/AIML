```python
#single inheritance
class Animal:
    def speak1(self):
        return "Animal speaks"
class Dog(Animal):
    def speak(self):
        return "Woof!"
dog=Dog()
print(dog.speak1())
print(dog.speak())

animal=Animal()
print(animal.speak1())
#print(animal.speak())
```

```python
#Multi inheritance
class Father:
    def skill(self):
        return "Gardening"
class Mother:
    def skill(self):
        return "Cooking"
class Child(Father, Mother):
    def skill(self):
        return Father.skill(self) + " and " + Mother.skill(self)

c=Child()
print(c.skill())
```

```python
#Multilevel Inheritance:
class Grandparent:
    def house(self):
        return "Owns a big house"

class Parent(Grandparent):
    def car(self):
        return "Own a car"

class Child(Parent):
    def bike(self):
        return "own a bike"

c=Child()
print(c.house())
print(c.car())
print(c.bike())
```

```python
#Hierarchical Inheritance:
class Parent:
    def property(self):
        return "Has land property"

class Son(Parent):
    pass
class Daughter(Parent):
    pass

s=Son()
d=Daughter()
print(s.property())
print(d.property())
```

```python
#### Hybrid Inheritance
class A:
    def method_A(self):
        return "Method from A"
class B(A):
    def method_B(self):
        return "Method from B"
class C(A):
    def method_C(self):
        return "Method from C"
class D(B, C):
    def method_D(self):
        return "Method from D"
obj = D()
print(obj.method_A())
print(obj.method_B())
print(obj.method_C())
print(obj.method_D())
```

```python
#Using super():
class Animal:
    def __init__(self,name):
        self.name= name
        print("Animal Created")

class Dog(Animal):
    def __init__(self,name,breed):
        super().__init__(name)
        self.breed=breed
        print("Dog created")

d=Dog("Buddy","Labrador")
print(d.name)
print(d.breed)
    
```

```python
#Class Methods and Static Methods:

class Employee:
    company = "XYZ Ltd"

    def __init__(self,name):
        self.name = name

    @classmethod
    def get_company(obj):
        return obj.company

    @staticmethod
    def greet():
        return "Welcome to the company!"

print(Employee.get_company())
print(Employee.greet())

    
```

```python
class Book:
    def __init__(self,title,pages):
        self.title=title
        self.pages=pages

    def __str__(self):
        return f"Book: {self.title},Pages:{self.pages}"

    def __len__(self):
        return self.pages

    def __mul__(self,other):
        return self.pages*other.pages


b1=Book("python",300)
b2=Book("Java",400)
print(b1)
print(len(b1))
print("Total Pages:",b1*b2)
```

```python
#Method Overloading:
class Math:
    def add(self,a,b,c=0):
        return a+b+c

obj=Math()
print(obj.add(2,3))
print(obj.add(2,3,4))
```

```python
#Operator Overloading
class Operator:
    def concat(self,d,e):
        return d+e
obj=Operator()
print(obj.concat("Abhy","nesh"))
print(obj.concat(2,3))
```

```python
class Parent:
    def greets(self):
        return "Hello from Parent"

class Child(Parent):
    def greet(self):
        return "Hello from Child"
        
obj=Child()
print(obj.greets())
```

```python
class Animal:
    def speak(self):
        print("Animal Speaks")
class Dog(Animal):
    def speak(self):
        print("Dog barks")
class Cat(Animal):
    def speak(self):
        print("Cat Meowsssssss")
obj1=Animal()
obj1.speak()
obj2=Dog()
obj2.speak()
obj3=Cat()
obj3.speak()
```

```python
#Composition:
class Engine:
    def start(self):
        return "Engine Started"
class Car:
    def __init__(self):
        self.engine=Engine()
    def drive(self):
        return self.engine.start() + "-> Car started moving"

c=Car()
print(c.drive())
```
