```python
#Importing libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataset:
dataset=pd.read_csv('Social_Network_Ads.csv')
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
dataset
```

```python
#Splitting training and testing set:
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
#Feature Scaling:
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
#Training the K-NN model on the Training set:
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
```

```python
#Predicting a new result:
print(classifier.predict(sc.transform([[27,100000]])))
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Confusion Matrix:
from sklearn.metrics import confusion_matrix,accuracy_score

cm=confusion_matrix(y_test,y_pred)
print(cm)
print("Testing Accuracy:",accuracy_score(y_test,y_pred))
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
#train data
y_pred_train=classifier.predict(X_train)
print("Traing Accuracy:",metrics.accuracy_score(y_train,y_pred_train))
print(classification_report(y_train,y_pred_train))
```

```python
#Visulaising the training set result:
from sklearn.inspection import DecisionBoundaryDisplay
# Plot Decision Boundary
DecisionBoundaryDisplay.from_estimator(
classifier,
X_train,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=dataset.columns[0],
        ylabel=dataset.columns[1],
    )
# Scatter plot
plt.scatter(X_train[:, 0], X_train[:, 1], 
            c=y_train, 
            s=20, edgecolors="k")
plt.title('KNN-Classifier(Training set)')
plt.show()
```

```python
#Visulaising the test set results:
from sklearn.inspection import DecisionBoundaryDisplay
# Plot Decision Boundary
DecisionBoundaryDisplay.from_estimator(
        classifier,
        X_test,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=dataset.columns[0],
        ylabel=dataset.columns[1],
    )
# Scatter plot
plt.scatter(X_test[:, 0], X_test[:, 1], 
            c=y_test, 
            s=20, edgecolors="k")
plt.title('KNN-Classifier (Testing set)')
plt.show()
```
