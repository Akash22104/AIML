```python
import sqlite3

conn=sqlite3.connect("my_database.db")
cursor=conn.cursor()
print("Database created and connected successfully!")
```

```python
#create a table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    grade TEXT,
    UNIQUE(name,age,grade)  -- prevents duplicate rows with same name+age+grade
)
""")

print("Table created successfully!")
```

```python
#Insert single record
cursor.execute("Insert into students(name,age,grade) VALUES (?,?,?)",
               ("Akash",22,"O"))

#Insert multiple records
students= [
    ("Bala",21,"A+"),
    ("Charlie",20,"B"),
    ("Delphine",23,"C"),
]
cursor.executemany("INSERT INTO students (name,age,grade) VALUES (?,?,?)",students)

conn.commit()

print("Data inserted successfully!")
```

```python
#Querying Data:
cursor.execute("SELECT * FROM students")
rows=cursor.fetchall()

print("All Students:")
for row in rows:
    print(row)
```

```python
#Filtering Data with WHERE
#Students with grade A
cursor.execute("SELECT * FROM students WHERE grade='C'")
rows=cursor.fetchall()

print("Students with grade C:")
for row in rows:
    print(row)
```

```python
#Students older than 21
cursor.execute("SELECT * FROM students WHERE age >21")
rows= cursor.fetchall()

print("Students older than 21:")
for row in rows:
    print(row)
#Students younger than 22
cursor.execute("SELECT *FROM students WHERE age < 22")
roes=cursor.fetchall()

print("\nStudents younger than 22:")
for row in rows:
    print(row)
```

```python
#LIKE(Pattern Matching)
cursor.execute("Select * from students WHERE name LIKE 'C%'")
rows=cursor.fetchall()

print("Students with names starting with 'C':")
for row in  rows:
    print(row)

#Names ending with 'a'
cursor.execute("SELECT * FROM students WHERE name LIKE '%a'")
rows=cursor.fetchall()

print("\nStudents with names ending in 'a':")
for row in rows:
    print(row)
    
```

```python
#IN (Multiple Values)
#Students with grade A or C:
cursor.execute("SELECT * FROM students WHERE grade IN ('A+','C')")
rows=cursor.fetchall()

print ("Students with grade A or C :")
for row in rows:
    print(row)
```

```python
#BETWEEN (Range)
#Studnets aged between 20 and 22(inclusive)
cursor.execute("SELECT * FROM students WHERE age BETWEEN 20 AND 22")
for row in rows:
    print(row)
```

```python
#Updating Records:
cursor.execute("UPDATE students  SET grade='B+' WHERE name='Akash'")
conn.commit()
```

```python
#Verify update:
cursor.execute("SELECT * FROM students WHERE name='Akash'")
print(cursor.fetchone())
```

```python
#Delecting Records:
cursor.execute("DELETE FROM students WHERE name='Delphine'")
conn.commit()
```
