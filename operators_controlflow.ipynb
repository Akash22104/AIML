```python
atm_pin=4141
balance=100000

while True:
    enter_pin =int(input("Enter your ATM PIN"))

    if enter_pin==atm_pin:
        print("PIN ACCEPTED")
        withdrawl = int(input("Enter the amount to be withdraw:"))

        if withdrawl <= balance:
            balance-=withdrawl
            print(f"Withdrawl successful! Remaining balance:₹{balance}")
        else:
            print(" Insufficient balance!")
        break
    else:
        print("wrong pin! please try again")
        
```
