```python
titanic_df = pd.read_csv('titanic.csv')
titanic_df.head(100)
```

```python
embarked = titanic_df['Embarked'].value_counts()
print(embarked.index) # these will be our bar labels
print(embarked.values) # these will be the heights of the bars
```

```python
embarked
```

```python
plt.bar(x=embarked.index,height=embarked.values);
```

```python
plt.bar(x=embarked.index, height=embarked.values, alpha= 0.5,color=['blue', 'green', 'red']);
```

```python
plt.barh(y=embarked.index, width=embarked.values, color=['blue', 'green', 'red']);
```

```python
embarked_m = titanic_df[titanic_df['Gender']=='male']['Embarked'].value_counts()
embarked_f = titanic_df[titanic_df['Gender']=='female']['Embarked'].value_counts()
print(embarked_m)
print(embarked_f)
```

```python
plt.bar(x=embarked_m.index, height=embarked_m.values, color='blue')
plt.bar(x=embarked_f.index, height=embarked_f.values, bottom=embarked_m.values, color='green')
```

```python
plt.bar(x=embarked_m.index, height=embarked_m.values, color='blue', width=0.75)
plt.bar(x=embarked_f.index, height=embarked_f.values, bottom=embarked_m.values, color='red', width=0.75)
plt.xlabel('Port')
plt.ylabel('Passengers')
plt.title('Port of Embarkation by Sex')
plt.legend(labels=['M', 'F']);
```

```python
age = titanic_df['Age'].dropna() # Drop missing values before plotting or the boxplot will be 
fig = plt.figure(figsize=(6, 6))
plt.boxplot(age, labels=['Age']);
```

```python
#3D plot
age = titanic_df['Age']
fare = titanic_df['Fare']
pclass = titanic_df['Pclass']
fig = plt.figure(figsize=(9, 9))
ax = plt.axes(projection='3d')
ax.scatter(xs=age, ys=fare, zs=pclass)
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Class');
```

```python
fig = plt.figure(figsize=(9, 9))
ax = plt.axes(projection='3d')
ax.scatter(xs=age, ys=fare, zs=pclass, c=pclass, cmap='brg')
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Class');
```

```python
#Subplot:
s_vals=titanic_df['Gender'].value_counts()
e_vals=titanic_df['Embarked'].value_counts()
fig,axes=plt.subplots(1,2)
fig.set_size_inches(8,4)
axes[0].pie(s_vals,labels=['male','Female'],startangle=90,autopct='%.1f%%')
axes[0].set_title('Gender');

axes[1].pie(e_vals, labels=['S', 'C', 'Q'], startangle=90, explode=[0, 0.1, 0])
axes[1].set_title('Embarked')
```

```python
fig,axes=plt.subplots(1,3)
fig.set_size_inches(12,4)

axes[0].pie(s_vals,labels=['Male','Females'],startangle=90,autopct='%.3f%%')
axes[0].set_title('Gender')
age=titanic_df['Age'].dropna()
axes[1].boxplot(age,labels=['Age']);

age=titanic_df['Age']
fare=titanic_df['Fare']
axes[2].scatter(x=age,y=fare);
```
